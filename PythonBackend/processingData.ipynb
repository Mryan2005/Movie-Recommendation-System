```python
import json
import pandas as pd
```

读取 douban_stories.json

```python
with open('douban_stories.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

查看数据结构

```python
print(type(data))
print(len(data))
print(type(data[0]))
```

将attributes内的这些部分独立出来

```python
key_mapping = {
    '导演': 'Director',
    '编剧': 'Writer',
    '主演': 'Cast',
    '类型': 'Genre',
    '制片国家/地区': 'Country_Region',
    '语言': 'Language',
    '上映日期': 'Release_Date',
    '首播': 'Premiere',
    '季数': 'Seasons',
    '集数': 'Episodes',
    '单集片长': 'Runtime',
    '片长': 'Runtime',
    '又名': 'AKA',
    'IMDb': 'IMDb'
}

for item in data:
    item.pop('官方小站', None)

    # 处理 attributes
    if 'attributes' in item and isinstance(item['attributes'], dict):
        attributes_dict = item.pop('attributes')
        new_attributes = {}

        for k, v in attributes_dict.items():
            en_key = key_mapping.get(k, k)

            # 使用拼接代替转化为JSON字符串，避免引发引号冲突
            if isinstance(v, list):
                v = ' / '.join([str(i) for i in v])
            elif isinstance(v, dict):
                v = json.dumps(v, ensure_ascii=False)

            new_attributes[en_key] = v

        item.update(new_attributes)

    # 处理 celebrities
    if 'celebrities' in item and isinstance(item['celebrities'], list):
        item['celebrities'] = {
            person['name']: json.dumps(person, ensure_ascii=False)
            for person in item['celebrities']
            if isinstance(person, dict) and 'name' in person
        }
```

```python
for item in data:
    if 'celebrities' in item and isinstance(item['celebrities'], list):
        item['celebrities'] = json.dumps(item['celebrities'], ensure_ascii=False)
```

```python
for item in data:
    release_date = item.get('Release_Date')
    premiere = item.get('Premiere')

    merged = []
    if release_date:
        merged.append(str(release_date))
    if premiere:
        merged.append(str(premiere))

    item['Release_Premiere'] = ' / '.join(merged) if merged else None

    item.pop('Release_Date', None)
    item.pop('Premiere', None)
```

使其变成pandas DataFrame

```python
df = pd.DataFrame(data)
```

```python
df = df[df['title'].notna() & (df['title'].astype(str).str.strip() != '')].reset_index(drop=True)
df = df.drop(columns=['官方小站'], errors='ignore')
```
